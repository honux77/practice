## 역삼각함수를 이용해서 PI 값을 구하는 프로그램

### 먼저 PI 값을 200자리까지 저장해 놓고 비교용으로 사용한다.

```python
from mpmath import mp
mp.dps = 102
PI = mp.pi
strPI = str(PI)
strPI = strPI[2:]
print(strPI)
print(len(strPI))
```

### 아크탄젠트 함수를 이용해서 PI 구하기

```python
def comparePI(s):
    c = 0
    v = str(s)[2:]    
    for i in range(101):
        if strPI[i] != v[i]: break
        c += 1
    print(c, str(s)[:100])
```

```python
def getPI(c):
    mp.dps = 1000000
    s = mp.mpf(0)    
    n = 1
    while True:        
        if n % 2 == 1:
            s += mp.mpf(1) / (mp.mpf(2) * n - mp.mpf(1))
        else:
            s -= mp.mpf(1) / (mp.mpf(2) * n - mp.mpf(1))
        if n % 1000000 == 0:
            print(n)
            comparePI(s * 4)
        if n == c:
            return s
        n += 1
```

```python
getPI(1000000000)
```
